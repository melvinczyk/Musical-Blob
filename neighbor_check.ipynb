```python
import json
import numpy as np

DATA_DIR = "./data"

with open(f"{DATA_DIR}/embeddings/latent_embeddings.json") as f:
    raw = json.load(f)

genres    = list(raw.keys())
token2idx = {g: i for i, g in enumerate(genres)}
Z         = np.array(list(raw.values()), dtype=np.float32)
Z_norm    = Z / (np.linalg.norm(Z, axis=1, keepdims=True) + 1e-9)

def neighbors(query, k=5):
    q = f"g_{query}" if not query.startswith("g_") else query
    if q not in token2idx:
        print(f"{q} not in vocab"); return
    idx  = token2idx[q]
    sims = Z_norm @ Z_norm[idx]
    top  = np.argsort(sims)[::-1][1:k+1]
    print(f"{q:30s} → {[genres[i] for i in top]}")

neighbors("pop rap")
neighbors("death metal")
neighbors("jazz")
neighbors("classical")
neighbors("edm")
neighbors("folk")
```
